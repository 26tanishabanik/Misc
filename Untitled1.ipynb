```python
data=data.drop(['DistArea_ID','Counterfeit_Weight'],axis=1)
```

```python
X_train=data.iloc[:,1:-1]
Y_train=data.iloc[:,-1]
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
X_train = pd.get_dummies(X_train,prefix_sep = '_',columns=['Medicine_Type','SidEffect_Level','Area_Type','Area_City_Type','Area_dist_level'])
print(X_train)
```

```python
X_train=X_train.values[:,:-1]
```

```python
from sklearn.preprocessing import Imputer
imputer=Imputer()
X_train=imputer.fit_transform(X_train)
print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
scale1=StandardScaler()
scaled1=scale1.fit_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
X_train.shape
Y_train=Y_train.values.reshape(6818,1)
```

```python
Y_train.shape
```

```python
X_train.shape
```

```python
plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('(Training set)')
plt.xlabel('Medicine_MRP')
plt.ylabel('Sales')
plt.show()
```
